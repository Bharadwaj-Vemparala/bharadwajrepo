```python
import  pandas as pd
import numpy as np
df=pd.read_csv('netflix.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
print("\nContent Type Distribution:")
print(df['type'].value_counts())
print("\nTop 10 Countries with Most Content:")
print(df['country'].value_counts().head(10))
print("\nTop 10 Content Ratings:")
print(df['rating'].value_counts().head(10))
print("\nMost Popular Genres:")
print(df['listed_in'].value_counts().head(10))
```

```python
categorical_columns = ["type", "rating", "country", "director", "cast", "listed_in"]
for col in categorical_columns:
    df[col] = df[col].astype("category")
```

```python
data_types = df.dtypes
print("Data Types:\n", data_types)
```

```python
data_types = df.dtypes
print("Data Types:\n", data_types)
```

```python
unique_values = df.nunique()
print("Unique Values:\n", unique_values)
```

```python
key_columns = ["director", "listed_in"]
for col in key_columns:
    print(f"\nUnique values in {col}:\n", df[col].unique()[:10])
```

```python
print("\nValue Counts for Categorical Attributes:")
for col in categorical_columns:
    print(f"\n{col} Value Counts:\n", df[col].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['release_year'], bins=20, kde=True)
plt.title("Distribution of Release Years")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type',color='red')
plt.title('Movies vs. TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
genres_over_time = df.groupby('release_year')['listed_in'].count()

plt.figure(figsize=(12,6))
plt.plot(genres_over_time.index, genres_over_time.values, marker='o', color='red', linestyle='-')
plt.xlabel('Release Year')
plt.ylabel('Count of Genres')
plt.title('Genre Trends Over Time')
plt.grid()
plt.show()
```

```python
top_directors = df['director'].value_counts().nlargest(10)

plt.figure(figsize=(12,6))
plt.barh(top_directors.index, top_directors.values, color='orange')
plt.xlabel('Number of Shows')
plt.ylabel('Director')
plt.title('Top 10 Directors with Most Shows')
plt.gca().invert_yaxis()  # Invert y-axis to show highest at top
plt.show()
```

```python
plt.figure(figsize=(18,14))
df['genres'] = df['listed_in'].str.split(',')
genres = df.explode('genres')
sns.countplot(y='genres', data=genres, order=genres['genres'].value_counts().index)
plt.title('Top Genres on Netflix')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month_added'] = df['date_added'].dt.month
monthly_counts = df['month_added'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
plt.bar(monthly_counts.index, monthly_counts.values, color='skyblue')
plt.xlabel("Month")
plt.ylabel("Number of Releases")
plt.title("Best Time to Release Content on Netflix")
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

This suggests that Netflix might focus on releasing more content during summer and holiday seasons when user engagement is typically higher.

If a country like India or South Korea has shown consistent growth, Netflix should invest more in content production for that region.

Genres like True Crime or Sci-Fi is gaining popularity, Netflix should increase investment in that genre.
